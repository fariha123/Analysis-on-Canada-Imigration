## Canda Imigration 

This project will be an analysis of the imigrating prople who are entering from differnt countries to Canada. We will analyze and find out which country has the highest numeber of imigrating people and the insights of the data.   

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import datetime
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import os

cwd = os.path.abspath('G:/PROJECTS/data analyst portfolio/canada imigration/') 
files = os.listdir(cwd) 
```

```python
df_total = pd.DataFrame()
for file in files:  # loop through Excel files
    if file.endswith('.xlsx'):
        excel_file = pd.ExcelFile('G:/PROJECTS/data analyst portfolio/canada imigration/Canada.xlsx')
        sheets = excel_file.sheet_names
        for sheet in sheets: # loop through sheets inside an Excel file
            df = excel_file.parse(sheet_name = sheet)
            df_total = df_total.append(df)
df_total.to_excel('combined_file.xlsx')
```
